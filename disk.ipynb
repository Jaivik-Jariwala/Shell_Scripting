```python
import matplotlib.pyplot as plt
```

```python
# FCFS disk scheduling
def fcfs(head, queue):
    total_movement = 0
    for i in range(len(queue)):
        diff = abs(head - queue[i])
        total_movement += diff
        head = queue[i]
    return total_movement
```

```python
# Sample disk queue and head position
queue = [86, 1470, 913, 1774, 948, 1509, 1022, 1750, 130]
head = 143
```

```python
 # Calculate total head movement for each algorithm
fcfs_total = fcfs(head, queue)
```

```python
print("FCFS total head movement:", fcfs_total)
```

```python
plt.bar(["FCFS", "SSTF", "SCAN", "C-SCAN", "LOOK", "C-LOOK"],
[fcfs_total])
plt.title("Disk Scheduling Algorithms")
plt.ylabel("Total Head Movement")
plt.show()
```
